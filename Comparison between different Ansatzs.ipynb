```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def get_E_range(filename, n):
    with open(filename, 'r') as infile:
        for i in range(1, n+1):
            infile.readline()
            line = infile.readline().split(' ')
    E_range = []
    for element in line:
        E_range.append(complex(element))
        
    return E_range
```

```python
h_range = np.linspace(0,2,20)
num_qubits = 2
Exact_E_range = get_E_range('measured_data/Exact_E_range.txt', num_qubits)
GeneralAnsatz_E_range = get_E_range('measured_data/GeneralAnsatz_E_range.txt', num_qubits)
RealAmpLinear_E_range = get_E_range('measured_data/RealAmpLinear_E_range.txt', num_qubits)
RealAmpFull_E_range = get_E_range('measured_data/RealAmpFull_E_range.txt', num_qubits)
RealAmpCir_E_range = get_E_range('measured_data/RealAmpCircular_E_range.txt', num_qubits)

plt.plot(h_range, Exact_E_range, label = 'Exact')
plt.plot(h_range, GeneralAnsatz_E_range, '^', label = 'GeneralAnsatz, Cobyla')
plt.plot(h_range, RealAmpFull_E_range, 's', label = 'RealAmp, full_etg, Cobyla')
plt.plot(h_range, RealAmpLinear_E_range, 'o', label = 'RealAmp, linear_etg, Cobyla')
plt.plot(h_range, RealAmpCir_E_range, '*', label = 'RealAmp, circular_etg, Cobyla')
plt.xlabel('h')
plt.ylabel('E/N')
plt.title('N = {}'.format(num_qubits))
plt.legend()
plt.show()
```
